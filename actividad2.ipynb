```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
cv2.__version__
```

```python
img = cv2.imread('./imagens/dog1.jpg')
imgShape = img.shape
print(imgShape)
```

```python
img = img[:,:,::-1]   #Esto porque matplotlib utiliza BGR en vez de RGB
plt.imshow(img)     
plt.show()
```

```python
edgesImg = cv2.Canny(img,100,200)  #Se aplica el filtro Canny con umbrales 100 y 200 y se almacena en edgesImg
plt.subplot(121),plt.imshow(img,cmap = 'gray') #Se hace una subgrfica de la imagen original
plt.title('Imagen original'), plt.xticks([]), plt.yticks([]) #Se establece el ttulo de la imagen y se especifca que no se ponga informacin en los ejes
plt.subplot(122),plt.imshow(edgesImg,cmap = 'gray') #Se hace una subgrfica de la imagen filtrada con los bordes resultantes del filtro Canny
plt.title('Imagen filtrada con los bordes'), plt.xticks([]), plt.yticks([]) #Se establece el ttulo de la imagen y se especifca que no se ponga informacin en los ejes

plt.show()
```

```python
cv2.imwrite('./imagens/dogEdges.jpg',edgesImg)
```

```python
imgGray = cv2.imread('imagens/dog1.jpg', cv2.IMREAD_GRAYSCALE)

# Aplicar el filtro de Harris para deteccin de esquinas
corner_img = cv2.cornerHarris(imgGray, blockSize=2, ksize=3, k=0.04)

# Dilatar la imagen de esquinas para hacerlas ms visibles
corner_img = cv2.dilate(corner_img, None)

# Preparar la imagen para visualizacin (normalizar)
corner_img_normalized = cv2.normalize(corner_img, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)

# Crear subgrficas
plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Imagen original'), plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(corner_img_normalized, cmap='gray')
plt.title('Deteccin de Esquinas (Harris)'), plt.xticks([]), plt.yticks([])

plt.show()
```
